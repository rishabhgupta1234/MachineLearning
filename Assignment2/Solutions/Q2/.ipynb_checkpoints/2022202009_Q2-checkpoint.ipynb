```python
idxTrain, idxTest = 0, 0
for idx, direc in enumerate(os.listdir(dirName)):
    path = os.path.join(dirName, direc)
    if os.path.isdir(path):
        i = 0
        for imgName in os.listdir(path):
            label, filePath = int(imgName.split('.')[0]), os.path.join(path, imgName)
            gray = cv2.imread(filePath, 0)
            gray = cv2.resize(gray, (112, 92))
            mat = np.asmatrix(gray).ravel()
            if i < 8:
                X_train[idxTrain,:] = mat
                y_train.append(idx)
                idxTrain += 1
            else:
                X_test[idxTest,:] = mat
                y_test.append(idx)
                idxTest += 1
            i += 1
```

#### Implement PCA Algorithm.

Explain your steps with comments and write a brief explanation of the method.

```python
"""
Implement pca in the following function by taking the following arguements:
1. X(images array)
2. number of components
The function must return the eigen vectors (principal components) also termed "Eigen faces"
"""
def principalComponentAnalysis(X, num_components):
    mean_face = np.mean(X, axis=0)
    X = X - mean_face
    covariance_mat = np.cov(X.T)
    eig_val, eig_vector = np.linalg.eigh(covariance_mat)
    sorted_indices = np.argsort(eig_val)[::-1]
    eig_vect = eig_vector[:,sorted_indices]
    return eig_vect[:,:num_components], eig_val, eig_vect, mean_face
```

```python
pca, eig_val, eig_vect, mean_face = principalComponentAnalysis(X_train, 50)
```

#### Implement Image Reconstruction from Eigenfaces


Explain your steps with comments and write a brief explanation of the method.

```python
# Function to implement image reconstruction by taking arguements as necessary.
def imageReconstruction(image, eigh_vect, mean):
    example_image = np.dot(image - mean, eigh_vect)
    return mean + np.dot(example_image, eigh_vect.T)
```

Visualise the results for different number of principal components(pc = 5,10,100, 1000, all).

#### Implement face recognition module based on the norm
Explain your steps with comments and write a brief explanation of the method.

*   Test the module and report accuracies based on the number of components taken for a range of value and plot them. 
*   Also plot the mean square error vs the number of eigenvectors taken and report your observations. 
*   For further empirical analysis, plot the semi-log variant of the error plot obtained above.

```python
"""
Implement pca in the following function by taking the following arguements:
1. Test image
2. Eigenvectors
3. Mean Eigen Face
4. Training Set(X_train,y_train)
The function must return the following:
1. image with smallest error
2. error value
3. prediction class
"""
def model(testImg, train_data, proj_test, y_train, eigen_vector):
    proj_train = np.dot(train_data, eigen_vector)
    diff = proj_train - testImg
    distances = np.sqrt(np.sum(diff ** 2, axis = 1))
    nearest_img = np.argmin(distances)
    return nearest_img, distances[nearest_img], y_train[nearest_img]
```

```python
"""
Iterate through all the images in the test data and 
test the accurate by taking different number of components
"""
mse = []
accuracies = []

for comp in n_components:
    v, pred, err = eig_vect[:,:comp], [], []
    proj_test = np.dot(X_test, v)
    for testImg in proj_test:
        predImg, error, predLabel = model(testImg, X_train, proj_test, y_train, v)
        diff = predImg - testImg
        pred.append(predLabel)
        err.append(diff)
    acc = accuracy_score(y_test, pred)
    accuracies.append(acc)
    mse.append(np.mean(np.square(err)))
    print("When component = %d, accuracy = %f" % (comp, acc))
```

```python
def plotGraph(x, y, xLabel, yLabel):
    plt.plot(x, y)
    plt.xlabel(xLabel)
    plt.ylabel(yLabel)
    plt.show()
```

```python
plotGraph(n_components, accuracies, 'Number of components', 'Accuracy')
```

```python
# Plot Number of eigenvectors vs Mean Square Error
plotGraph(n_components, mse, 'Number of components', 'MSE')
```

```python
# Plot Number of eigenvectors vs Logarithmic Mean Square Error
plotGraph(n_components, np.log(mse), 'Number of components', 'Logarithmic MSE')
```

State the advantages and limitations of using PCA in general. Also state its limitations in context of face recognition

**Answer:**
Advantages of PCA:-
- It can be used for dimensionality reduction.
- It can be used for feature extraction.

Disadvantanges of PCA:-
- Loss of information
- Computationally expensive
- Sensitive to outliers

In context of face recognition, the limitations of PCA can be:-
- Sensitive to the structure of image (pose)
- Might give poor performance in case of bad lighting conditions
