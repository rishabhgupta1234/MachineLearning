## Part1: K Means

```python
#implement elbow method from scratch
def elbow(X, y, centroids):
 
  sqSum = 0
  n=X.shape[0]
  
  for it1 in range(0,n):
    distArray = []
    c=np.unique(y).size
    for it2 in range(c):
      distArray.append(np.linalg.norm(np.array(X[it1])-np.array(centroids[it2]), 2))
    sqSum =sqSum+ np.min(np.array(distArray))

  return sqSum
```

```python
# LOADING THE DATASET. Take only the test split (10000 samples)
from keras.datasets import fashion_mnist

(trainX, trainy), (testX, testy) = fashion_mnist.load_data()
X = testX.reshape(testX.shape[0] , -1)
y = np.copy(testy)
print("10K images flattened: " , X.shape , " and 10k labels: " , y.shape)
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import Normalizer

normalizer = Normalizer()
X = normalizer.fit_transform(X)
```

KMeans (Kmeans++ default)

```python
K = list(range(2, 21))
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

elbowValues0 = []
silhouetteScore0 = []
 
for k in K:
  kmeans = KMeans(k, init="k-means++", random_state=1)
  kmeans.fit(X)
  elbowValues0.append(elbow(X, kmeans.predict(X), kmeans.cluster_centers_))
  silhouetteScore0.append(silhouette_score(X, kmeans.predict(X)))
```

```python
print(elbowValues0)
print(silhouetteScore0)
```

```python
plt.title('Elbow Score vs K values')
plt.xlabel('K values')
plt.ylabel('Elbow Score Values')
plt.plot(K, elbowValues0)
```

```python
plt.title('Silhouette Score vs K values')
plt.xlabel('K values')
plt.ylabel('Silhouette Score Values')
plt.plot(K, silhouetteScore0)
```

Best K is 3 for k-means++

1B.) Forgy (random in sklearn)

```python
elbowValues1 = []
silhouetteScore1 = []
 
for k in K:
  kmeans = KMeans(k, init='random', random_state=1)
  kmeans.fit(X)
  elbowValues1.append(elbow(X, kmeans.predict(X), kmeans.cluster_centers_))
  silhouetteScore1.append(silhouette_score(X, kmeans.predict(X)))
```

```python
print(elbowValues1)
print(silhouetteScore1)
```

```python
plt.title('Elbow Score vs K values')
plt.xlabel('K values')
plt.ylabel('Elbow Score Values')
plt.plot(K, elbowValues1)
```

```python
plt.title('Silhouette Score vs K values')
plt.xlabel('K values')
plt.ylabel('Silhouette Score Values')
plt.plot(K, silhouetteScore1)
```

Best K is 3 for random kmeans

1C.) Visualise the dataset to depict the clusters formed (prefer T-SNE)


```python
from sklearn.manifold import TSNE

X = TSNE(n_components=2, n_jobs=-1).fit_transform(X)
```

```python
# Fit KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=1).fit(X)

plt.figure(figsize=(10, 8))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('TSNE Visualization of K-Means Clustering')
plt.xlabel('TSNE Feature 1')
plt.ylabel('TSNE Feature 2')
plt.colorbar()
plt.show()
```

```python
# Fit KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3, init='random', random_state=1).fit(X)
 
plt.figure(figsize=(10, 8))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('TSNE Visualization of K-Means Clustering')
plt.xlabel('TSNE Feature 1')
plt.ylabel('TSNE Feature 2')
plt.colorbar()
plt.show()
```

```python
#visualize a dendogram
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(X, method='ward')
plt.figure(figsize=(15, 10))
dendrogram(Z, truncate_mode='level', p=10)
plt.show()


  
```

**Question**: Explain the procedure to find the optimal number of clusters using dendogram

Answer: </br>
</br>
The optimal number of clusters using dendrogram is 2.

It is equal to the number of points which we get after drawing intersecting horizontal line on the max vertical height which is not intersecting with other vertical heights.




```python
# Single linkage clustering
singleClustering = AgglomerativeClustering(n_clusters=2, linkage='single',affinity="euclidean")
singleClustering.fit(X)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(X[:, 0], X[:, 1], c=singleClustering.labels_,cmap="rainbow")
plt.title('Clusters of Single Linkage')
plt.show()
```

```python
# Complete linkage clustering
completeClustering = AgglomerativeClustering(n_clusters=2, linkage='complete',affinity="euclidean")
completeClustering.fit(X)
```

```python
#visualisation of clusters using single linkage and complete linkage

# Plot the clusters obtained using complete linkage
plt.figure(figsize=(10, 7))
plt.scatter(X[:, 0], X[:, 1], c=completeClustering.labels_,cmap="rainbow")
plt.title('Clusters of Complete Linkage')
plt.show()

```

**Question:** Mention the differences observed using single and complete linkage:

Answer: </br>
- Complete linkage considers the maximum distance between all pairs of points in two clusters, while single linkage only considers the minimum distance between them.
- In our case single linkage is giving more seperated clusters in comparison to complete linkage,because it taking only nearby points not farther points.
- Single linkage is more sensitive to noise and outliers in the data than complete linkage in our plot

```python
# Boiler plate code
import cv2

image = cv2.imread('dog.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# cv2_imshow(img)
img_size = image.shape

# Reshape it to be 2-dimension
pixels = image.reshape(img_size[0] * img_size[1], img_size[2])
```

```python

pixels = image.reshape((-1, 3))
```

```python

import math
from skimage.io import imsave, imread ,imshow

Kvalues = [2,5,9,13,17]
for k in Kvalues:

    kmeans = KMeans(n_clusters=k, random_state=1).fit(pixels)
    compressedPixels = kmeans.cluster_centers_[kmeans.labels_]
    compressedImage = compressedPixels.reshape(image.shape)
    myimage = np.array(compressedImage, dtype=np.uint8)
    imsave(f"compressedImage_k{k}.jpg", myimage)
    originalSize = image.size * 24
    compressedSize = k * 24 + image.size * math.ceil(math.log2(k))
    dataReduction = (originalSize - compressedSize)/originalSize
    dataReductionPercentage=dataReduction * 100
    dataReductionPercentage =round(dataReductionPercentage, 2)
    print('Data reduction percentage for', k, 'is :',dataReductionPercentage)
```
